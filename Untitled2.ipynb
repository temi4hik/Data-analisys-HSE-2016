```python
digits = load_digits()
images = digits.images
```

```python
def apply_kernel(image, kernel):
    i_size = image.shape[0]
    
    bordered_image = np.zeros((i_size + 2, i_size + 2))
    
    results = np.zeros((i_size, i_size))
    
    
    for i in range(1, i_size + 1):
        bordered_image[i,:][1 : i_size + 1] = image[i - 1,:]
    
    for i in range(i_size):
        for j in range(i_size):
            square = bordered_image[i:i+3,j:j+3]
            s = 0
            for d in range(3):
                for n in range(3):
                    s += kernel[d][n] * square[d][n]
            results[i][j] = s
            
    return results
```

```python
k = np.array([[-1, -1, -1], [0, 0, 0], [1, 1, 1]])
k1 = np.array([[-1, -2, -1], [0, 0, 0], [1, 2, 1]])


d_x = apply_kernel(images[1], k1)
d_y = apply_kernel(images[1], k1.T)
```

```python
img_der = (d_y ** 2 + d_x ** 2) ** (0.5)
```
