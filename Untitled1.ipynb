```python
import numpy as np
import numpy.random as npr
import matplotlib.pyplot as plt
from itertools import izip
import sys
%matplotlib inline

D_TYPES = ('EUC', 'L1', 'CHEB', 'MINK', 'EUC_SQR')
```

```python
mu1 = np.array([1., 1.])
mu2 = np.array([3., 4.])
mu3 = np.array([-2., 4.])

S1 = np.eye(2)
S2 = np.eye(2)
S3 = np.eye(2)

X1 = npr.multivariate_normal(mu1, S1, 100)
X2 = npr.multivariate_normal(mu2, S2, 100)
X3 = npr.multivariate_normal(mu3, S3, 100)


X = np.concatenate([X1, X2, X3])
# mu = np.array([0., 0.])
# S = np.eye(2)

# X = npr.multivariate_normal(mu, S, 300)
```

```python
def get_distance(x, y, metric='EUC', pw=3):
    res = 0.0
    if metric == D_TYPES[0]:
        res = sum([(x_i - y_i) ** 2 for x_i, y_i in izip(x, y)]) ** (0.5)
    elif metric == D_TYPES[1]:
        res = sum([abs(x_i - y_i) for x_i, y_i in izip(x, y)])
    elif metric == D_TYPES[2]:
        res = max([abs(x_i - y_i) for x_i, y_i in izip(x, y)])
    elif metric == D_TYPES[3]:
        res = sum([(x_i - y_i) ** pw for x_i, y_i in izip(x, y)]) ** (1.0 / pw)
    elif metric == D_TYPES[4]:
        return get_distance(x, y) ** 2
    
    return float(res)
```

```python
CONDITIONS = ('iters', 'z_change', 'mu_change')
```

```python
def find_clasters(X, g, init_randomly=False, condition='iters', iter_num=1, epsilon=0.01, metric='EUC', pw=3):
    size = X.shape[0]
    clasters = []
    
    centers = init_centers(X, g, metric, pw, randomly=init_randomly)
    centers_old = np.zeros((g, 2))
    
    Z = np.zeros(size)
    Z_old = np.ones(size)
    
    cond = None
    j = 0
    
    if condition == 'iters':
        cond = lambda: j < iter_num
    elif condition == 'z_change':
        cond = lambda: (Z == Z_old).all()
    elif condition == 'mu_change':
        cond = lambda: abs(centers - centers_old).max() < epsilon
    
    while not cond():
        i = 0
        Z_old = np.copy(Z)
        centers_old = np.copy(centers)
        
        for x in X:
            new_z_i = get_min_idx(x, centers, metric, pw)
            old_z_i = Z[i]

            if new_z_i != old_z_i:
                Z[i] = new_z_i
                centers[old_z_i] = get_new_center(X, Z, old_z_i)
                centers[new_z_i] = get_new_center(X, Z, new_z_i)
            i += 1
        
        clasters.append([np.copy(Z), np.copy(centers)])
        j += 1
    return clasters


def get_new_center(points, Z, z_i):
    c = np.zeros(2)
    
    pts = points[(Z == z_i)]
    
    for p in pts:
        c += p
    c /= pts.shape[0]
    
    return c

    
def init_centers(points, g, metric, pw, randomly=False):
    
    if randomly:
        indices = npr.randint(points.shape[0], size=g)
        return points[indices]

    mu = np.zeros(2)
    for p in points:
        mu += p
    mu /= points.shape[0]
    
    mu_1 = get_center([mu], points, metric, pw)
    
    mus = [mu_1]
    
    for i in xrange(1, g):
        mu_i = get_center(mus, points, metric, pw)
        mus.append(mu_i)
    return np.array(mus)
        
def get_center(centers, points, metric, pw):
    max_dist = 0
    max_idx = 0
    for c in centers:
        max_dist += get_distance(c, points[0], metric, pw)
    
    N = points.shape[0]
    
    for i in xrange(1, N):
        new_dist = sum([get_distance(c, points[i], metric, pw) for c in centers])
        
        if new_dist > max_dist:
            max_dist = new_dist
            max_idx = i
    
    return points[max_idx]


def get_min_idx(x, points, metric, pw):
    min_dist = get_distance(x, points[0], metric, pw)
    min_idx = 0
    
    N = points.shape[0]
    for i in xrange(1, N):
        dist = get_distance(x, points[i], metric, pw)
        if dist < min_dist:
            min_dist = dist
            min_idx = i
            
    return min_idx
```

```python
cls_num = 3
iter_num = 2

clasters = find_clasters(X, cls_num, init_randomly=True, iter_num=iter_num, condition='mu_change', 
                         metric='EUC_SQR')

colors = np.array(['y', 'm', 'c', 'r', 'g', 'b', 'k'])
# colors = np.array(['b', 'r'])
npr.shuffle(colors)

it = 0

for cls in clasters:
    plt.figure(it)
    
    clasts, centers = clasters[it]
    plt.title('With {0} iterations'.format(it + 1))
    for i in range(cls_num):
        X_i = X[(clasts == i)]
        plt.scatter(X_i[:,0], X_i[:,1], c=colors[i])
        plt.plot([centers[i][0]], [centers[i][1]], 'x', c=colors[i], mew=3, ms=20)
    
    plt.show()
    it += 1
```
